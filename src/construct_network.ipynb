```python
import networkx as nx
```

```python
list_people = [('piyush', 'chiara'), ('chiara', 'sara'), ('sanjo', 'sara'), ('sanjo', 'piyush')]
```

```python
nodes = [
            {1: 'piyus'},
            {2: 'chiara'},
            {3: 'sanjo'},
            {4: 'sara'}
]
```

```python
edges = [(1,2), (2,4), (3,4), (3,1)]
```

```python
G0 = nx.DiGraph(edges)
```

```python
nx.draw(G0)
```

```python
G0.number_of_edges()
G0.number_of_nodes()
```

```python
G = nx.DiGraph(list_people)
```

```python
nx.draw(
    G, 
    with_labels=True, 
    node_size=7000,
    node_color= '#EDD382', 
    node_shape='h',
    width = 5,
    edge_color = '#FF521B',
    
    )
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(15, 9))
ax.axis('off')
plot_options = {"node_size": 10, "with_labels": False, "width": 0.15}
nx.draw_networkx(G, pos=nx.random_layout(G), ax=ax, **plot_options)
```

```python
import networkx as nx
# import matplotlib.style as style, matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout
import dzcnapy_plotlib as dzcnapy

F = nx.DiGraph()
F.add_node("C")
F.add_edges_from([("B", "b0"), ("b0", "b1"), ("b1", "B")])
F.add_edges_from([("A", "a0"), ("a0", "a1"), ("a1", "a2"), ("a1", "a3"),
                  ("a3", "A")])
pos = graphviz_layout(F)
nx.draw_networkx(F, pos, **dzcnapy.attrs)
dzcnapy.set_extent(pos, plt)
dzcnapy.plot("abcNetwork")
```
